## Simple Descriptive analysis

The First thing we can notice is that there is no mean, median, std deviation, miniumum, maximum or quantiles (25-50-75 percentiles).

- top : Indicates the most frequently occurring value within each column.
- frequency : Represents the frequency or count of the top value in each column
